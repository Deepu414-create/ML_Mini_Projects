```python
"""Project Title: Student Performance Prediction

Summary:
This project demonstrates a simple Linear Regression model in Python to predict student scores based on the number of study hours. The dataset is created manually in CSV format. The workflow includes:
Loading dataset with pandas
Training a Linear Regression model
Predicting student scores
Visualizing data and regression line using Matplotlib
This mini project helps in understanding the basics of supervised machine learning (regression) and data handling with CSV files."""


import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("students.csv")
print("Dataset:\n")
print(df)
print("\n Dataset Info:")
print(df.mean(numeric_only=True))
df.plot(x="Name",kind="bar",figsize=(8,5))
plt.title("Student Marks Comparison")
plt.ylabel("Marks")
plt.show()


```
